```python
# face alignment
from imutils.face_utils import FaceAligner
from imutils.face_utils import rect_to_bb
import numpy as np
import imutils
import dlib
import cv2
import os

detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
fa = FaceAligner(predictor, desiredFaceWidth=100)
img_path = "cropped"
files = os.listdir(img_path)
for i in files:
    print(i)
    img = cv2.imread(img_path + '/' + i)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = detector(gray, 1)
    print(i)
    j = 0
    for face in faces:
        (x, y, w, h) = rect_to_bb(face)
        faceOrig = imutils.resize(img[y : y + h, x : x + w], width=100)
        faceAligned = fa.align(img, gray, face)
        cv2.imwrite('./face_org/{}_{}'.format(j, i), faceOrig)
        cv2.imwrite('./face_aligned/{}_{}'.format(j, i), faceAligned)
        j += 1
```

```python
from imutils import face_utils
import dlib
import cv2
import os
import pandas as pd
 
# Vamos inicializar um detector de faces (HOG) para então
# let's go code an faces detector(HOG) and after detect the 
# landmarks on this detected face

# p = our pre-treined model directory, on my case, it's on the same script's diretory.
p = "shape_predictor_68_face_landmarks.dat"
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(p)

img_path = "face_aligned"
files = os.listdir(img_path)

dic = {}
x_values = [[] for _ in range(48)]
y_values = [[] for _ in range(48)]
labels = []
for i in files:
    label = i[-5:][0]
    if label == '0' or label== '3':
        image = dlib.load_rgb_image(img_path + '/' + i)
        dets = detector(image, 1) 
        num_face = len(dets)
        print(i, num_face)
        if num_face == 1:
            labels.append(int(label))
            for k, d in enumerate(dets):
    #         print("Detection {}: Left: {} Top: {} Right: {} Bottom: {}".format(
    #             k, d.left(), d.top(), d.right(), d.bottom()))
                # Get the landmarks/parts for the face in box d.
                shape = predictor(image, d)
                for j in range(48):
                #             print(i, shape.part(i).x, shape.part(i).y)
                    x_values[j].append(shape.part(j).x)
                    y_values[j].append(shape.part(j).y)

for i in range(48):
    dic['x' + str(i + 1)] = x_values[i]
    dic['y' + str(i + 1)] = y_values[i]

dic['label'] = labels
df = pd.DataFrame(data=dic)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

df_y = df['label'] == 3 
df_X = df[['x' + str(i) for i in range(1,49)] + ['y' + str(j) for j in range(1,49)]]
X_train, X_test, y_train, y_test = train_test_split(df_X,df_y,test_size = 0.2,random_state=15)

# Feature Scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()  
X_train = sc.fit_transform(X_train)  
X_test = sc.transform(X_test)

clf = RandomForestClassifier(n_estimators=10)
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,y_pred))  
print(classification_report(y_test,y_pred))  
print(accuracy_score(y_test, y_pred))
```
